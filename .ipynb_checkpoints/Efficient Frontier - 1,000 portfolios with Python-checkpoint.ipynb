```python
df = pd.read_csv('Markowitz_Data.csv', index_col = 'Date')
```

#### need log returns first 

```python
log_returns = np.log(df/df.shift(1))
```

```python
log_returns.mean() * 250 #to get annual returns
```

```python
log_returns.cov() * 250
```

```python
log_returns.corr()
```

```python
num_assets = len(df.columns)
```

#### need to create random weights

```python
arr = np.random.random(2)
arr
```

```python
arr[0] + arr[1]
```

#### can't use the above because it won't always sum to 1 
need to use another method below 

```python
weights = np.random.random(num_assets)
weights /= np.sum(weights) # adjusts weights to sum = 1
weights
```

```python
weights[0] + weights[1]
```

### Expected Portfolio Return

```python
np.sum(weights * log_returns.mean()) * 250 #annualize the expected portfolio returns, using the weights and log returns
```

### Expected Portfolio Variance

```python
np.dot(weights.T, np.dot(log_returns.cov() * 250, weights))
```

### Expected Volatility 

```python
np.sqrt(np.dot(weights.T, np.dot(log_returns.cov() * 250, weights)))
```

#### going to run 1,000 simuations of the portfolio with different weights, ranging from 99% PG -1% S&P500 and vice versa

```python
#this is going to build a graph and shows the hypothetical returns versus the volitilites, i.e. we need two objects to store the data

pfolio_returns = [] #going to a list together of random returns
pfolio_vol = []

for x in range(1000):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    pfolio_returns.append(np.sum(weights * log_returns.mean()) * 250)
    pfolio_vol.append(np.sqrt(np.dot(weights.T, np.dot(log_returns.cov() * 250, weights))))

pfolio_returns, pfolio_vol    
```

#### However, the lists above are not useful so we will convert to numpy arrays 

```python
pfolio_returns = [] #going to a list together of random returns
pfolio_vol = []

for x in range(1000):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    pfolio_returns.append(np.sum(weights * log_returns.mean()) * 250)
    pfolio_vol.append(np.sqrt(np.dot(weights.T, np.dot(log_returns.cov() * 250, weights))))

pfolio_returns = np.array(pfolio_returns)
pfolio_vol = np.array(pfolio_vol)
    
pfolio_returns, pfolio_vol    
```

#### Need to create a dataframe to put the arrays into

```python
portfolios = pd.DataFrame({"Return": pfolio_returns, "Volatility": pfolio_vol})
```

```python
portfolios.head()
```

```python
portfolios.tail()
```

```python
portfolios.plot(x="Volatility", y="Return", kind= 'scatter', figsize=(10,6))
plt.title('Efficient Frontier (n=1,000)')
plt.xlabel("Expected Volatility")
plt.ylabel("Expected Return")
```
