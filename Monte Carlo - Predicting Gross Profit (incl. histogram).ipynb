# Sample analysis: Forcasting Gross Profit

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170 #revenue mean
rev_stdev = 20 #revenue stdev
iterations = 1000 #creating a normal distribution
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev #created 1,000 values 
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show() 
```

#### Almost all of our data points fall within 150 and 190

```python
COGS = - (rev * np.random.normal(0.6,0.1)) #COGS 60% of revenue - using past data, and a stdev of 10%, negative. We have 1,000 revenue data points and need a COGS for each of those points 

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
gross_profit = rev + COGS
gross_profit
```

```python
plt.figure(figsize=(15,6))
plt.plot(gross_profit)
plt.show()
```
